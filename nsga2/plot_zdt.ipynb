# Plots dos problemas ZDT e fronteiras de Pareto geradas pelo NSGA-II

Este notebook mostra os plots das fronteiras Pareto-Ótimas dos problemas ZDT, assim como os melhores indivíduos não dominados gerados pelo algoritmo NSGA-II (http://www.iitk.ac.in/kangal/codes.shtml).

Os dados dos indivíduos estão nos arquivos plot_*.out gerados pelo código do link acima.

```python
import numpy as np

%matplotlib inline
import matplotlib.pyplot as plt
```

## ZDT1

```python
data = np.loadtxt("Resultados/plot_zdt1.out").T
```

```python
fig = plt.figure(figsize=(8,6))

# fronteira pareto-ótima
x = np.arange(0, 1, 0.01);
y = 1 - np.sqrt(x)
plt.plot(x, y, '-', label='Fronteira Pareto-Ótima')

# dados do NSGA-II
plt.plot(data[0], data[1], '.', label='População')

plt.legend()

plt.show()
```

## ZDT2

## ZDT3

```python
data = np.loadtxt("Resultados/plot_zdt3.out").T
```

```python
fig = plt.figure(figsize=(8,6))

# fronteira pareto-ótima
x = np.arange(0, 1, 0.01);
y = 1 - np.sqrt(x) - x * np.sin(10*np.pi*x)
plt.plot(x, y, '-', label='Função para g(x) = 1')

# dados do NSGA-II
plt.plot(data[0], data[1], '.', label='População')

plt.legend()

plt.show()
```
